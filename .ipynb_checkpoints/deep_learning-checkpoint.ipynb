```python
from tensorflow import keras
from tensorflow.keras import layers

from tensorflow.keras.preprocessing import image_dataset_from_directory
```

```python
IMG_WIDTH = 299
IMG_HEIGHT = 299
CHANNELS = 3
DIR = './dataset/'
NCLASS = 7

train = image_dataset_from_directory(
    DIR,
    image_size=(IMG_WIDTH, IMG_HEIGHT),
    validation_split=0.3,
    label_mode='categorical',
    subset='training',
    batch_size=32,
    seed=42)

val = image_dataset_from_directory(
    DIR,
    image_size=(IMG_WIDTH, IMG_HEIGHT),
    validation_split=0.3,
    label_mode='categorical',
    subset='validation',
    batch_size=32,
    seed=42)
```

```python
data_augmentation = keras.Sequential(
    [layers.RandomFlip("horizontal"), layers.RandomRotation(0.1),]
)

# part 1: xception
base_model = keras.applications.Xception(
    weights='imagenet',
    input_shape=(IMG_WIDTH, IMG_HEIGHT, CHANNELS),
    include_top=False
)

base_model.trainable = False

inputs = keras.Input(shape=(IMG_WIDTH, IMG_HEIGHT, CHANNELS))
x = data_augmentation(inputs)  # Apply random data augmentation
scale_layer = keras.layers.Rescaling(scale=1 / 127.5, offset=-1)
x = scale_layer(x)
x = base_model(x, training=False)
x = keras.layers.GlobalAveragePooling2D()(x)
x = keras.layers.Dropout(0.2)(x)  # Regularize with dropout
outputs = keras.layers.Dense(NCLASS)(x)
model = keras.Model(inputs, outputs)
model.compile(
    optimizer=keras.optimizers.Adam(),
    loss=keras.losses.CategoricalCrossentropy(),
    metrics=[keras.metrics.CategoricalAccuracy()],
)
epochs = 20
history_xception = model.fit(train, epochs=epochs, validation_data=val)
```

```python
base_model.trainable = True
model.summary()

model.compile(
    optimizer=keras.optimizers.Adam(1e-5),  # Low learning rate
    loss=keras.losses.BinaryCrossentropy(),
    metrics=[keras.metrics.BinaryAccuracy()],
)

epochs = 10
history_ft_xception = model.fit(train, epochs=epochs, validation_data=val)
```

```python
IMG_WIDTH = 224
IMG_HEIGHT = 224

train = image_dataset_from_directory(
    DIR,
    image_size=(IMG_WIDTH, IMG_HEIGHT),
    validation_split=0.3,
    label_mode='binary',
    subset='training',
    batch_size=32,
    seed=42)

val = image_dataset_from_directory(
    DIR,
    image_size=(IMG_WIDTH, IMG_HEIGHT),
    validation_split=0.3,
    label_mode='binary',
    subset='validation',
    batch_size=32,
    seed=42)

# part 2: mobilenet
base_model = keras.applications.MobileNetV3Small(
    weights='imagenet',
    input_shape=(IMG_WIDTH, IMG_HEIGHT, CHANNELS),
    include_top=False
)

base_model.trainable = False

inputs = keras.Input(shape=(IMG_WIDTH, IMG_HEIGHT, CHANNELS))
x = data_augmentation(inputs)  # Apply random data augmentation
scale_layer = keras.layers.Rescaling(scale=1 / 127.5, offset=-1)
x = scale_layer(x)
x = base_model(x, training=False)
x = keras.layers.GlobalAveragePooling2D()(x)
x = keras.layers.Dropout(0.2)(x)  # Regularize with dropout
outputs = keras.layers.Dense(NCLASS)(x)
model = keras.Model(inputs, outputs)
model.compile(
    optimizer=keras.optimizers.Adam(),
    loss=keras.losses.BinaryCrossentropy(),
    metrics=[keras.metrics.BinaryAccuracy()],
)
epochs = 20
history_mobilenet = model.fit(train, epochs=epochs, validation_data=val)

base_model.trainable = True
model.summary()

model.compile(
    optimizer=keras.optimizers.Adam(1e-5),  # Low learning rate
    loss=keras.losses.BinaryCrossentropy(),
    metrics=[keras.metrics.BinaryAccuracy()],
)

epochs = 10
history_ft_mobilenet = model.fit(train, epochs=epochs, validation_data=val)
```

```python
# part 3: nasnet
base_model = keras.applications.NASNetMobile(
    weights='imagenet',
    input_shape=(IMG_WIDTH, IMG_HEIGHT, CHANNELS),
    include_top=False
)

base_model.trainable = False

inputs = keras.Input(shape=(IMG_WIDTH, IMG_HEIGHT, CHANNELS))
x = data_augmentation(inputs)  # Apply random data augmentation
scale_layer = keras.layers.Rescaling(scale=1 / 127.5, offset=-1)
x = scale_layer(x)
x = base_model(x, training=False)
x = keras.layers.GlobalAveragePooling2D()(x)
x = keras.layers.Dropout(0.2)(x)  # Regularize with dropout
outputs = keras.layers.Dense(NCLASS)(x)
model = keras.Model(inputs, outputs)
model.compile(
    optimizer=keras.optimizers.Adam(),
    loss=keras.losses.BinaryCrossentropy(),
    metrics=[keras.metrics.BinaryAccuracy()],
)
epochs = 20
history_nasnet = model.fit(train, epochs=epochs, validation_data=val)

base_model.trainable = True
model.summary()

model.compile(
    optimizer=keras.optimizers.Adam(1e-5),  # Low learning rate
    loss=keras.losses.BinaryCrossentropy(),
    metrics=[keras.metrics.BinaryAccuracy()],
)

epochs = 10
history_ft_nasnet = model.fit(train, epochs=epochs, validation_data=val)
```

```python
import matplotlib.pyplot as plt

plt.plot(
    history_xception.epoch,
    [el * 100 for el in history_xception.history['val_loss']],
    label='Xception')
plt.plot(
    history_mobilenet.epoch,
    [el * 100 for el in history_mobilenet.history['val_loss']],
    label='MobileNetV3')
plt.plot(
    history_nasnet.epoch,
    [el * 100 for el in history_nasnet.history['val_loss']],
    label='NASNetMobile')
plt.grid()
plt.legend(loc=4)
title = DIR.split('/')[-1]
plt.title(f'Validation accuracy on {title}-vs-all')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.savefig(f'{DIR}.png', dpi=600)
```
